```python
from collections import defaultdict
```

```python
chapter_dictionary = defaultdict(list)
chapter_number = 0
```

```python
for line in lines:
    word_list = line.split( )
    if word_list:
        first_word = word_list[0]
        if first_word[-1] == '.':
            try:
                float(first_word)
                chapter_number += 1
            except:
                pass
        
        chapter_dictionary[chapter_number].append(line)
```
