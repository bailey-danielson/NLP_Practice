```python
import pandas as pd
import re
from nltk.corpus import stopwords
```

```python
full_df = pd.read_csv("/Users/BaileyDanielson/Documents/Python/NLP_Practice/final_df.csv")
```

```python
full_df.head()
```

```python
word_list = []
index_list = []
book_list = []
```

```python
stopwords = set(stopwords.words('english'))
stopwords.update(["said", "say" , "come", "one" , "tell" , "must", "would", "could"])
```

```python
for index, row in full_df.iterrows():
    for word in row.Sentence.split():
        new_word = word.lower()
        if new_word not in stopwords:
            clean_word = re.sub(r'[^a-z ]+', '', new_word)
            word_list.append(clean_word)
            index_list.append(row.Index)
            book_list.append(row.Book_Title)
```

```python
print(len(word_list), len(index_list), len(book_list))
```

```python
words_df = pd.DataFrame(
    {'Word_List': word_list,
     'Index_List': index_list,
     'Book_List': book_list
    })
```
